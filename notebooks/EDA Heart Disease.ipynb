### 1) Background:
- Cardiovascular diseases are a leading cause of mortality worldwide, with heart disease being a significant contributor to premature deaths. Early detection and accurate prediction of heart disease can enable timely medical interventions, improving patient outcomes and reducing healthcare costs. The  dataset, contains a comprehensive set of features related to demographic, lifestyle, and clinical factors that may influence the likelihood of heart disease. These features include age, gender, blood pressure, cholesterol levels, exercise habits, smoking status, family history of heart disease, diabetes, BMI, and various biomarkers such as triglyceride levels, fasting blood sugar, CRP, and homocysteine levels, along with the target variable indicating heart disease status.


### 2) Problem statement
- The objective of this machine learning project is to develop a predictive model that accurately identifies individuals at risk of heart disease based on the features provided in the "heart_disease.csv" dataset. The model will classify individuals into two categories: those with heart disease (Yes) and those without heart disease (No). By leveraging demographic, lifestyle, and clinical data, the model aims to assist healthcare professionals in early diagnosis, risk stratification, and personalized treatment planning for patients. The challenge lies in handling a mix of numerical (e.g., age, blood pressure, cholesterol levels) and categorical (e.g., gender, smoking status, exercise habits) features, addressing potential missing or noisy data, and selecting the most relevant features to improve model performance while avoiding overfitting.


### 3) Data Collection
- Dataset Source - https://www.kaggle.com/datasets/oktayrdeki/heart-disease
- The data consists of 21 column and 10000 rows.

### 4) Expected Impact
- The successful development of this predictive model will enable early identification of individuals at high risk of heart disease, facilitating preventive measures and targeted interventions. This can lead to improved patient outcomes, reduced healthcare costs, and a better understanding of the risk factors associated with heart disease, ultimately contributing to public health advancements.

### 2.2 Dataset information

### 4. Exploring Data ( Visualization )
#### 4.1 Visualize average score distribution to make some conclusion.

#### 4.1.1 Histogram 

```python
# Visualization 1: Histogram of Age Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="Age", bins=10, color="#ea3453")
plt.title("Age Distribution", fontsize=16)
plt.xlabel("Age (Years)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.tight_layout()
plt.show()
```

### 4.1.2 BarChart

```python
# Visualization 2: Bar Chart for Smoking vs Heart Disease
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x="Smoking", hue="Heart Disease Status", palette={"Yes": "#ef4444", "No": "#10b981"})
plt.title("Smoking and Heart Disease", fontsize=16)
plt.xlabel("Smoking Status", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Heart Disease")
plt.tight_layout()
plt.show()
```

```python
# Visualization 4: Box Plot for Sleep Hours vs Heart Disease (Interesting Fact)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Heart Disease Status", y="Sleep Hours", palette={"Yes": "#ef4444", "No": "#10b981"})
plt.title("Sleep Hours by Heart Disease Status", fontsize=16)
plt.xlabel("Heart Disease Status", fontsize=12)
plt.ylabel("Sleep Hours", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Visualization 5: Interactive Scatter Plot with Plotly
# We'll create a scatter plot of Age vs Cholesterol Level, colored by Heart Disease Status
import plotly.express as px
fig = px.scatter(
    df,
    x="Age",
    y="Cholesterol Level",
    color="Heart Disease Status",
    color_discrete_map={"Yes": "#ef4444", "No": "#10b981"},
    title="Interactive Scatter Plot: Age vs Cholesterol Level",
    labels={"Age": "Age (Years)", "Cholesterol Level": "Cholesterol Level (mg/dL)"},
    hover_data=["BMI", "Fasting Blood Sugar"],
)
fig.update_layout(
    width=800,
    height=600,
    title_font_size=16,
    xaxis_title_font_size=12,
    yaxis_title_font_size=12,
)
fig.show()
```

## Conclusion:
